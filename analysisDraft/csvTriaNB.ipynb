Simple trial analysis from CSV.

```python
import numpy as np
import pandas as pd

# try plotly instead of matplotlib
import plotly.tools as tools
import plotly.offline as py
import plotly.graph_objs as go
import plotly.figure_factory as ff
py.init_notebook_mode(connected=True)
```

```python
aa="/Users/cad/Desktop/dasd/12182017/trialData/dasd_0006_12182017_s16_trial_45.csv"
tempMeta=pd.read_csv(aa)
```

```python
sInds=["v1","rv1","counter1"]
dRange=[0,1000]
xV = tempMeta['psData.tm'][dRange[0]:dRange[1]]
y1 = tempMeta['psData.v1'][dRange[0]:dRange[1]]
y2 = tempMeta['psData.rv1'][dRange[0]:dRange[1]]
y3 = tempMeta['psData.rv3'][dRange[0]:dRange[1]]


# Create traces
trace0 = go.Scatter(
    x = xV,
    y = y1,
    mode = 'lines',
    name = sInds[0]
)
trace1 = go.Scatter(
    x = xV,
    y = y2,
    mode = 'lines',
    name = 'rv1'
)
trace2 = go.Scatter(
    x = xV,
    y = y3,
    mode = 'lines',
    name = 'rv3'
)


fig = tools.make_subplots(rows=2, cols=1,subplot_titles=("{} & {}".format(sInds[0],sInds[1]),sInds[2]));
fig.append_trace(trace0, 1, 1);
fig.append_trace(trace1, 1, 1);
fig.append_trace(trace2, 2, 1);
fig['layout'].update(height=600, width=600); # title='trial data'
py.iplot(fig, filename='simple-subplot',image='svg');
```
